```python
import tensorflow as tf
from tensorflow.keras import layers
import tensorflow_datasets as tfds
```

```python
(train_data, test_data), info = tfds.load(
    "imdb_reviews",
    split=["train", "test"],
    as_supervised=True,
    with_info=True
)
```

```python
BUFFER_SIZE = 10000
BATCH_SIZE = 64
VOCAB_SIZE = 10000
EMBEDDING_DIM = 64
MAX_LENGTH = 200
```

```python
encoder = tf.keras.layers.TextVectorization(
    max_tokens=VOCAB_SIZE,
    output_mode="int",
    output_sequence_length=MAX_LENGTH
)
```

```python
train_text = train_data.map(lambda text, label: text)
encoder.adapt(train_text)

# Encode and batch datasets
def preprocess(text, label):
    return encoder(text), label
```

```python
train_ds = train_data.map(preprocess).shuffle(BUFFER_SIZE).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_ds = test_data.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
model = tf.keras.Sequential([
    layers.Embedding(VOCAB_SIZE, EMBEDDING_DIM),
    layers.Bidirectional(layers.LSTM(64)),
model = tf.keras.Sequential([
    layers.Embedding(VOCAB_SIZE, EMBEDDING_DIM),
    layers.Bidirectional(layers.LSTM(64)),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history = model.fit(train_ds, epochs=5, validation_data=test_ds)

# Evaluate
loss, accuracy = model.evaluate(test_ds)
print(f"\nTest Accuracy: {accuracy:.4f}")
```
